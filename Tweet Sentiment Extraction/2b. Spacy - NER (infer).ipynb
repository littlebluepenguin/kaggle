Because this is a code competition, all submission must be made in kernels. This is a runner for the Spacy NER model

```python
import numpy as np
import pandas as pd

import spacy
```

```python
test = pd.read_csv('../input/tweet-sentiment-extraction/test.csv')
```

```python
def predict(nlp, texts):
    docs = nlp.pipe(texts)
    preds = []
    for i, doc in enumerate(docs):
        if doc.ents:
            text = doc.ents[0].text
            preds.append(text)
        else:
            preds.append(texts[i])
    return preds
```

```python
def save_submission(tweet_extracts):
    submission = pd.read_csv('../input/tweet-sentiment-extraction/sample_submission.csv')
    submission["selected_text"] = pd.Series(tweet_extracts).map(lambda x: f'\"{x}\"')
    submission.to_csv(f"submission.csv", index=False)
```

```python
nlp = spacy.load(f'../input/tweet-sentiment-extraction-2a-spacy-ner-train/ner_models/spacy')
preds = predict(nlp, test['text'])
save_submission(preds)
```
