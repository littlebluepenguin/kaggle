```python
import numpy as np
import pandas as pd

from utils import reduce_mem_usage

import pickle
```

```python
test = pd.read_csv('../input/feature-engineering/test_features.csv')
train = pd.read_csv('../input/feature-engineering/train_features.csv')
```

```python
train = reduce_mem_usage(train)
test = reduce_mem_usage(test)
```

```python
X, y = train.drop(['open_channels'], axis=1).values, train['open_channels'].values
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import f1_score

extra_trees = ExtraTreesRegressor(n_estimators=200,
                                max_features='sqrt',
                                random_state=0,
                                max_depth=25,
                                n_jobs=-1)

extra_trees.fit(X, y)

preds = extra_trees.predict(X)
preds = np.round(np.clip(preds, 0, 10))

f1_score(y, preds, average='macro')
```

```python
with open('extra_trees.pkl', 'wb') as f:
    pickle.dump(extra_trees, f)
```
